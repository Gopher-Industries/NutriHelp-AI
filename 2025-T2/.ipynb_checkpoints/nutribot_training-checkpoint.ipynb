```python
import numpy as np
import chromadb
import os
from groq import Groq
from dotenv import load_dotenv

load_dotenv()
```

```python
class RAG:
    def __init__(self, collection_name="aus_food_nutrition"):
        self.chroma_client = chromadb.PersistentClient(path=".chroma")
        self.collection = self.chroma_client.get_or_create_collection(name=collection_name)
        self.count = self.collection.count()

    def add_documents(self, docs):
        for doc in docs:
            _id = f"id{self.count}"
            self.collection.upsert(ids=[_id], documents=[doc])
            self.count += 1

    def retrieve(self, prompt, n_results=5):
        return self.collection.query(
            query_texts=[prompt],
            n_results=n_results
        )
```

```python
rag = RAG()
rag.add_documents([
    "Vegemite is a popular Australian spread made from brewers' yeast extract.",
    "Kangaroo meat is a lean source of protein, low in fat.",
])
```

```python
# Retrieve by prompt
analyzed_health_condition = """
{
    "obesity_prediction": {
        "obesity_level": "Overweight_Level_II"
        "confidence": 10%
    },
    "diabetes_prediction": {
        "diabetes": true,
        "confidence": 90%
    }
"""

weekly_plan_format = """{
"suggestion": STRING
"weekly_plan": [
    {
        "week": 1,
        "target_calories_per_day": INT,
        "focus": STRING,
        "workouts": [ARRAY OF STRINGS],
        "meal_notes": STRING,
        "reminders": [ARRAY OF STRINGS]
    },
    ... (repeat for as many weeks as appropriate)
}]"""
prompt = f"""
You are a nutrition and fitness assistant.
Below is an analyzed health condition for a user, expressed in JSON: {analyzed_health_condition}
Your task: Based on the analyzed health condition and using the retrieved knowledge, generate a weekly plan strictly in this JSON format (replace INT and STRING placeholders): {weekly_plan_format} 
"""
relevant_texts = rag.retrieve(prompt, n_results=2)
prompt = prompt + f"\tUse this as context for answering: {relevant_texts}"
print("Retrieved texts:", relevant_texts)
```

```python
def keyword_score(query, doc):
    query_words = set(query.lower().split())
    doc_words = set(doc.lower().split())
    return len(query_words & doc_words)

def best_chunks(query, initial_k=10, final_k=4):
    rag = RAG()
    res = rag.retrieve(query, n_results=initial_k)

    docs = res.get("documents", [[]])[0]
    distances = res.get("distances", [[]])[0] if res.get("distances") else [999] * len(docs)

    ranked = []
    seen = set()

    for i, doc in enumerate(docs):
        clean_doc = doc.strip()

        if not clean_doc:
            continue

        if clean_doc in seen:
            continue
        seen.add(clean_doc)

        dist = distances[i] if i < len(distances) else 999
        overlap = keyword_score(query, clean_doc)

        ranked.append({
            "doc": clean_doc,
            "distance": dist,
            "overlap": overlap
        })

    ranked = sorted(
    ranked,
    key=lambda x: (x["distance"], -x["overlap"], len(x["doc"]))
)

    return ranked[:final_k]
```

```python
def inspect_best_chunks(query, initial_k=10, final_k=4):
    best_chunk = best_chunks(query, initial_k=initial_k, final_k=final_k)

    print(f"\nQuery: {query}")
    print("=" * 80)

    for i, item in enumerate(best_chunk, 1):
        print(f"\nSelected Chunk {i}")
        print(f"distance = {item['distance']:.4f}")
        print(f"keyword overlap = {item['overlap']}")
        print(item["doc"])
        print("-" * 80)

inspect_best_chunks("What proportion of Australian children aged 2 to 17 do not eat enough vegetables?")
inspect_best_chunks("Which section covers actions to reduce harmful ingredients in processed foods?")
inspect_best_chunks("Name one 2025 resource related to the Healthy Food Partnership.")
```

```python
import os
import random
import re
from groq import Groq
from datasets import load_dataset
from dotenv import load_dotenv

# try loading .env
load_dotenv(dotenv_path=".env", override=True)

# fallback: set key manually here for testing only
# os.environ["GROQ_API_KEY"] = "your_real_key_here"

print("Current folder:", os.getcwd())
print("Does .env exist here?", os.path.exists(".env"))
print("Key loaded?", os.environ.get("GROQ_API_KEY") is not None)

class APIUnavailable(RuntimeError):
    pass


def _extract_item(example):
    q = example.get("Question") or example.get("question")
    ca = example.get("Correct Answer") or example.get("correct_answer")

    inc = [
        example.get("Incorrect Answer 1"),
        example.get("Incorrect Answer 2"),
        example.get("Incorrect Answer 3"),
    ]

    if not any(inc) and "incorrect_answers" in example:
        inc = example["incorrect_answers"]

    inc = [x.strip() for x in inc if isinstance(x, str) and x.strip()]
    return q, ca, inc


def evaluate_gpqa(question, options, model="llama-3.1-8b-instant"):
    api_key = os.environ.get("GROQ_API_KEY")
    if not api_key:
        raise RuntimeError("GROQ_API_KEY is missing from environment variables.")

    client = Groq(api_key=api_key)

    opts_str = "\n".join(f"{chr(65+i)}. {opt}" for i, opt in enumerate(options))

    prompt = (
        "You are answering a multiple-choice question.\n"
        "Choose the correct answer from the options below.\n\n"
        f"Question: {question}\n\n"
        f"Options:\n{opts_str}\n\n"
        "Return only the answer in this exact format:\n"
        "- ANSWER=<LETTER>"
    )

    try:
        resp = client.chat.completions.create(
            messages=[
                {
                    "role": "system",
                    "content": (
                        "You are a multiple-choice grader. "
                        "Return the answer in EXACTLY this format and nothing else:\n"
                       
                    ),
                },
                {"role": "user", "content": prompt},
            ],
            model=model,
            temperature=0,
        )

        ans = (resp.choices[0].message.content or "").strip()
        print(ans)

        m = re.search(r"ANSWER=([A-D])", ans)
        return m.group(1).upper() if m else None

    except Exception as e:
        print(f"API error: {e}")
        raise APIUnavailable from e


def run_eval(num_questions):
    try:
        dataset = load_dataset("Idavidrein/gpqa", "gpqa_diamond", split="train")
    except Exception as e:
        print("Dataset loading error:", e)
        return

    correct = 0
    attempted = 0

    for i in range(min(num_questions, len(dataset))):
        q, ca, inc = _extract_item(dataset[i])

        if not (q and ca and len(inc) >= 3):
            print(f"Skipping Q{i+1}: missing fields")
            continue

        options = inc[:3] + [ca]
        random.shuffle(options)
        true = chr(65 + options.index(ca))

        try:
            pred = evaluate_gpqa(q, options)
        except APIUnavailable:
            print("Stopping because API is unavailable.")
            break

        attempted += 1
        if pred == true:
            correct += 1

        print(f"Q{i+1}: {'✓' if pred == true else '✗'}  Pred={pred or '-'}  True={true}")

    if attempted == 0:
        print("No questions evaluated.")
    else:
        pct = 100.0 * correct / attempted
        print(f"\nFinal Score: {correct}/{attempted} ({pct:.1f}%)")


if __name__ == "__main__":
    run_eval(10)
```

# Customized QA Evaluation

```python
import os
from groq import Groq


def improve_retrieval(user_question, retrieved_items, top_k=3):
    """
    Rank retrieved chunks using simple runtime heuristics:
    - prefer smaller Chroma distance
    - reward keyword overlap with the user question
    - keep the final top_k chunks
    """

    question_words = set(
        word.strip(".,?!:;()[]{}'\"").lower()
        for word in user_question.split()
        if len(word.strip()) > 2
    )

    ranked = []

    for item in retrieved_items:
        doc = item.get("doc", "").strip()
        distance = item.get("distance", None)

        if not doc:
            continue

        doc_lower = doc.lower()

        overlap = sum(1 for word in question_words if word in doc_lower)

        # Smaller distance is better. If missing, give weaker default.
        distance_score = 0 if distance is None else (1 / (1 + distance))

        # Combined score: distance + keyword overlap
        score = distance_score + (0.15 * overlap)

        ranked.append({
            "doc": doc,
            "distance": distance,
            "overlap": overlap,
            "score": score
        })

    ranked.sort(key=lambda x: x["score"], reverse=True)
    return ranked[:top_k]


def chat_with_rag(user_question):
    rag = RAG()

    api_key = os.environ.get("GROQ_API_KEY")
    if not api_key:
        raise RuntimeError("GROQ_API_KEY missing")

    client = Groq(api_key=api_key)

    # Clean query before retrieval
    clean_question = user_question.replace("Keep concise", "").strip()

    # Retrieve from Chroma
    results = rag.retrieve(clean_question, n_results=10)

    retrieved_items = []

    if results:
        raw_docs = results.get("documents", [[]])[0]
        raw_distances = results.get("distances", [[]])[0] if "distances" in results else []

        for i, doc in enumerate(raw_docs):
            if not doc:
                continue

            doc = doc.strip()

            # Filter weak / noisy chunks
            if len(doc) <= 50:
                continue

            distance = raw_distances[i] if i < len(raw_distances) else None

            retrieved_items.append({
                "doc": doc,
                "distance": distance
            })

    # Remove exact duplicate chunks
    deduped_items = []
    seen_docs = set()

    for item in retrieved_items:
        if item["doc"] not in seen_docs:
            seen_docs.add(item["doc"])
            deduped_items.append(item)

    # Re-rank best chunks for runtime context
    selected_chunks = improve_retrieval(clean_question, deduped_items, top_k=3)

    # Build final context
    if selected_chunks:
        context_text = "\n\n".join(item["doc"] for item in selected_chunks)
    else:
        context_text = "No relevant context found."

    final_prompt = f"""
You are a helpful NutriHelp assistant.

Use only the retrieved context to answer the question.

Rules:
- Return exactly ONE short sentence
- Give only the most direct answer
- Do not explain or add extra details
- Do not restate the question
- Do not write "Answer:"
- If the question asks for a section or page name, return only the section or page name
- If the question asks for one resource, return only the resource name
- If the answer includes a statistic, include the full statistic in one sentence
- If the answer is not supported by the context, say exactly:
I could not find the answer in the provided context.

Examples:
94% of children aged 2 to 17 don't eat enough vegetables.
Only 1 in 13 adults eat enough fruit and vegetables.
About one-third of total energy intake comes from discretionary foods.
On the Food and nutrition topic page under "Eating what your body needs."
The "Reducing harmful ingredients" section in the Food and nutrition topic.
Healthy Food Partnership Program – Executive Committee communique – February 2025.

Context:
{context_text}

Question:
{clean_question}

Answer:
"""

    try:
        chat_completion = client.chat.completions.create(
            messages=[{"role": "user", "content": final_prompt}],
            model="llama-3.1-8b-instant",
            temperature=0
        )

        answer = chat_completion.choices[0].message.content.strip()

        if answer.lower().startswith("answer:"):
            answer = answer[len("answer:"):].strip()

        if not answer:
            answer = "I could not find the answer in the provided context."

        return answer.replace("’", "'")

    except Exception as e:
        return f"Error: {str(e)}"
```

```python
questions = [
    "What proportion of Australian children aged 2 to 17 do not eat enough vegetables?",
    "Roughly what proportion of Australian adults eat enough fruit and vegetables?",
    "About what share of Australia’s total energy intake comes from foods we don’t need?",
    "Where on the department website can I read about getting the nutrients I need, at any age?",
    "Which section covers actions to reduce harmful ingredients in processed foods?",
    "Name one 2025 resource related to the Healthy Food Partnership."
]

for q in questions:
    print("=" * 70)
    print(f"QUESTION: {q}")
    answer = chat_with_rag(q)
    print(f"ANSWER: {answer}")
    print()
```

# Add RAG from Crawler

```python
import json
import os

rag = RAG()
docs = []

file_path = "document-parser/data/sentences.jsonl"

if not os.path.exists(file_path):
    print("File not found:", file_path)
else:
    with open(file_path, "r", encoding="utf-8") as f:
        for line_number, line in enumerate(f, start=1):
            try:
                rec = json.loads(line)
                sent = rec.get("sentence", "").strip()
                if sent:
                    docs.append(sent)
            except json.JSONDecodeError:
                print(f"Skipping invalid JSON on line {line_number}")

    print("Number of documents loaded:", len(docs))

    # docs = docs[:1000]   # optional for testing

    rag.add_documents(docs)
    print("Documents added successfully.")
```
